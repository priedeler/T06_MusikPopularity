# Musik Popularity Prediction: Summary of Insights

This notebook provides a comprehensive overview of the results obtained from various machine learning models used to predict song popularity. We compare the effectiveness of different approaches, from simple audio-based classification to advanced NLP-enhanced models.

## 1. Project Overview

The goal of this project was to determine if we can predict whether a song becomes a **"hit"** (defined as a popularity score > 50) based on:
1.  **Audio Features**: Metadata like danceability, energy, tempo, and loudness.
2.  **Lyrics**: The textual content of the song, processed using Natural Language Processing (NLP).

We explored several models, including k-Nearest Neighbors (kNN), Logistic Regression, Random Forest, and Naive Bayes.

## 2. Comparing Audio-Only Models

First, we looked at how well models perform using only the standard Spotify audio features. This represents our baseline predictive capability.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Data gathered from previous analysis notebooks
audio_models_data = {
    'Model': ['kNN', 'Logistic Regression', 'Random Forest', 'Gaussian NB', 'Categorical NB (9-bins)'],
    'Accuracy': [0.65, 0.66, 0.68, 0.79, 0.81]
}

df_audio = pd.DataFrame(audio_models_data)

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
ax = sns.barplot(x='Model', y='Accuracy', data=df_audio, palette='viridis')

plt.title('Accuracy Comparison: Audio-Only Models', fontsize=15)
plt.ylim(0.5, 0.9)
plt.ylabel('Test Accuracy')

# Adding values on top
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

### Insights from Audio Features:
*   **Naive Bayes** variants performed surprisingly well, with the **Categorical NB** (using 9-bin discretization) achieving the highest accuracy (~81%) among pure audio models.
*   Among standard classifiers, **Random Forest** outperformed kNN and Logistic Regression, suggesting that non-linear decision boundaries and ensemble methods are better suited for this data.

## 3. The "Secret Sauce": Adding Lyrics

The most significant breakthrough occurred when we combined audio features with lyrical content. We used TF-IDF to vectorize the words and SVD (Singular Value Decomposition) to reduce dimensionality to the most important "topics" in the lyrics.

```python
improvement_data = {
    'Approach': ['Audio Only (RF)', 'Audio + Lyrics (RF)'],
    'Accuracy': [0.68, 0.89]
}

df_imp = pd.DataFrame(improvement_data)

plt.figure(figsize=(8, 6))
ax = sns.barplot(x='Approach', y='Accuracy', data=df_imp, palette='magma')

plt.title('Impact of Adding Lyrics to Prediction', fontsize=15)
plt.ylim(0.5, 1.0)
plt.ylabel('Test Accuracy')

for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

### Insights from Lyrics Analysis:
*   **Massive Improvement**: Adding lyrics boosted accuracy from **68% to 89%**. This indicates that while the "sound" of a song is important, the "message" or vocabulary used is a much stronger predictor of popularity in our dataset.
*   **TF-IDF + SVD**: This combination proved effective at capturing the essence of lyrics without overwhelming the model with thousands of sparse word features.

## 4. Key Feature Takeaways

Based on our various analyses, here are the key factors that contribute to a song's popularity:

| Feature Type | High Impact Factors | Observation |
| :--- | :--- | :--- |
| **Audio** | Danceability & Energy | Hit songs consistently show higher median scores in these features. |
| **Audio** | Loudness | Hits tend to be slightly louder on average. |
| **Lyrics** | Genre Context | Lyrics are highly descriptive of genre, which in turn correlates with popularity. |
| **Lyrics** | Topic Modeling | Certain "word topics" identified via SVD showed high correlation with hits. |

## 5. Final Conclusions

1.  **Simple models have limits**: Audio features alone can only get us to ~68-80% accuracy. There is a inherent "ceiling" when only considering the sound measurements.
2.  **Context is King**: The addition of lyrics provides the necessary context to break through the performance ceiling, reaching near 90% accuracy.
3.  **Preprocessing is Critical**: Discretizing continuous features (as seen in Naive Bayes) and robustly cleaning text (as seen in Lyrics Analysis) are just as important as the choice of model.
4.  **Ensemble Power**: Random Forest consistently provided robust results, especially when handling high-dimensional data like combined audio and text features.
